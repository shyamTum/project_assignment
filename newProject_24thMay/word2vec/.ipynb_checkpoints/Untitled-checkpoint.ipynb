```python
import pandas as pd
import nltk
nltk.download('punkt')
```

```python
df = pd.read_csv('simpsons_dataset.csv')
```

```python
df.head(10)
```

```python
print(df.isna().sum())
```

```python
df.shape
```

```python
df.dropna(how='any', inplace=True)
#df.fillna(df.mean())
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
df.shape
```

```python
df_spoken_words = df['spoken_words'].values
```

```python
df_spoken_words
```

```python
from gensim.models.phrases import Phrases, Phraser
```

```python
sent = [row.split() for row in df_spoken_words]
```

```python
sent
```

```python
# I tried to use flatten list. But not used in later steps.
flat_list = [item for sublist in sent for item in sublist]
phrases = Phrases(sent, min_count=20, progress_per=10000)
```

```python
flat_list
```

```python
bigram = Phraser(phrases)
sentences = bigram[sent]
```

```python
sentences
```

```python
#data = [nltk.word_tokenize(spoken_item) for spoken_item in df_spoken_words]
# sentences = df_spoken_words.toList()
```

```python
from gensim.models import Word2Vec
model = Word2Vec(vector_size=50, window=4, min_count=1, workers=4, sg=1)
```

```python
#model.train(df_spoken_words, total_examples=model.corpus_count, epochs=30, report_delay=1)
model.build_vocab(sentences, progress_per=10000)
```

```python
model.train(sentences, total_examples=model.corpus_count, epochs=30, report_delay=1)
```

```python
#model.wv.init_sims()
model.wv.get_vector('apple')
```

```python
model.wv.similarity('apple', 'pear')
```

```python
model.wv.similarity('apple','orange')
```

```python
model.wv.similarity('man','boy')
```

```python
model.wv.similarity('man','girl')
```

```python
model.wv.most_similar('apple')
```

```python
model.wv.most_similar('pear')
```
